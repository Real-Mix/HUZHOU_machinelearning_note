# 线性回归
* ## model：
$$ Y = \beta x+ \epsilon$$

$$\begin{bmatrix}{y_1}\\{y_2}\\{\vdots}\\{y_m}\end{bmatrix}=
\begin{bmatrix}
{x_{11}}&{x_{12}}&{\cdots}&{x_{1n}}\\
{x_{21}}&{x_{22}}&{\cdots}&{x_{2n}}\\
{\vdots}&{\vdots}&{\ddots}&{\vdots}\\
{x_{m1}}&{x_{m2}}&{\cdots}&{x_{mn}}\\
\end{bmatrix}
\begin{bmatrix}{x_1}\\{x_2}\\{\vdots}\\{x_m}\end{bmatrix}+
\begin{bmatrix}{\epsilon_1}\\{\epsilon_2}\\{\vdots}\\{\epsilon_m}\end{bmatrix}
$$

* ## strategy：
平方损失 $E(f)=\frac{1}{n}\sum_{i=1}^N(y_i-\hat y)^2 =\frac{1}{n}\sum_{i=1}^N(\epsilon)^2 $

* ## algorithm ：
微分法——>最小二乘法得到**显式解**
$$\hat \beta _{LS}=(X^TX)^{-1}X^TY$$

# 逻辑斯提分布
设 $X$ 是连续随机变量， $X$ 服从逻辑斯谛分布是指 $X$ 具有下列分布函数和密度函数:
$$
\begin{gathered}
F(x)=P(X \leq x)=\frac{1}{1+e^{-(x-\mu) / \gamma}} \\
f(x)=F^{\prime}(x)=\frac{e^{-(x-\mu) / \gamma}}{\gamma\left(1+e^{-(x-\mu) / \gamma}\right)^2}
\end{gathered}
$$
式中， $\mu$ 为位置参数， $\gamma>0$ 为形状参数。分布函数曲线以点 $\left(\mu, \frac{1}{2}\right)$ 为中心对称，曲线在中心附近增长速度较快，在两端增长速度较慢。形状参数 $\gamma$ 的值越小，曲线在中心附近增长得越快。

# 问题：
* 已知 $Data=\{{(x_i,y_i)},i=1,2,..,N\}$，$x_i \in R^n$， $ y_i\in\{0,1\}$为分成的类别数，对于新的实例 $ x\in R^n$，如何确定 $y$ ?

# model


* 根据  $Data=\{{(x_i,y_i)},i=1,2,..,N\}$ 学习出 $w$ ，使得
$$ Y=y=  \underset{y}{argmax}P(Y=1|x)=\frac{exp(w·x)}{exp(w·x)+1}$$


# strategy & algorithm：——> MLE(极大似然估计)

逻辑斯谛回归模型学习时, 对于给定的训练数据集 $T=\left\{\left(x_1, y_1\right),\left(x_2, y_2\right), \cdots\right.$, $\left.\left(x_N, y_N\right)\right\}$, 其中, $x_i \in \mathbf{R}^n, y_i \in\{0,1\}$, 可以应用极大似然估计法估计模型参数, 从而得到逻辑斯谛回归模型。

设：
$$
P(Y=1 \mid x)=\pi(x), \quad P(Y=0 \mid x)=1-\pi(x)
$$
似然函数为
$$
\prod_{i=1}^N\left[\pi\left(x_i\right)\right]^{y_i}\left[1-\pi\left(x_i\right)\right]^{1-y_i}
$$
对数似然函数为
$$
\begin{aligned}
L(w) &=\sum_{i=1}^N\left[y_i \log \pi\left(x_i\right)+\left(1-y_i\right) \log \left(1-\pi\left(x_i\right)\right)\right] \\
&=\sum_{i=1}^N\left[y_i \log \frac{\pi\left(x_i\right)}{1-\pi\left(x_i\right)}+\log \left(1-\pi\left(x_i\right)\right)\right] \\
&=\sum_{i=1}^N\left[y_i\left(w \cdot x_i\right)-\log \left(1+\exp \left(w \cdot x_i\right)\right]\right.
\end{aligned}
$$
对 $L(w)$ 求极大值, 得到 $w$ 的估计值。

* 求出**梯度** （对数似然函数在该点的**一阶**导数）：$$g=\frac{\partial lnL(w)}{\partial w}= \sum_{i=1}^{n}[y_i-\frac{exp(w·x_i)}{1+exp(w·x_i)}]x_i $$

* 求出**黑塞矩阵**（对数似然函数在该点的**二阶**导数）：$$H=\frac{\partial^2 lnL(w)}{\partial w^2}= -[\frac{exp(w·x_i)}{1+exp(w·x_i)}][\frac{1}{1+exp(w·x_i)}]x_i^Tx_i $$

* 使用**牛顿法**：得到迭代公式为 $$w^{(k+1)}=w^{(k)}-H^{-1}g$$
迭代到收敛停止或人工设定停止迭代的次数

# 牛顿迭代法逻辑斯蒂回归代码实现

```python
import numpy as np
import pandas as pd

def logistic(x):
    return 1/(1+np.exp(-x))


###########################################
w = np.zeros(2)
w[0] = 1
w[1] = -1

x_train = pd.DataFrame({'x1': np.random.uniform(-5, 5, 50), 'x2': np.ones(50)})
y_train = np.array([np.random.binomial(1, logistic(
    np.dot(w, x_train.iloc[i, :])), 1) for i in range(x_train.shape[0])])

############################################
eps = 1e-4
eta = 5*1e-2

k = 0
errval = 0
w_0 = np.random.uniform(0, 1, 2)
max_it = 500
w_n = w_0

for k in range(max_it):
    fx = np.array(sum([(y_train[i]-logistic(np.dot(w_n, x_train.iloc[i, :])))
                  * x_train.iloc[i, :] for i in range(x_train.shape[0])]))
    dfx = -sum([logistic(np.dot(w_n, x_train.iloc[i, :]))*(1-logistic(np.dot(w_n, x_train.iloc[i, :])))
               * np.dot(x_train.iloc[i, :], x_train.iloc[i, :]) for i in range(x_train.shape[0])])
    w_n = w_n-eta*fx/dfx
    errval = np.sum([(y_train[i]-logistic(np.dot(w_n, x_train.iloc[i, :])))
                    ** 2 for i in range(x_train.shape[0])])/50

    #print("%3d [%22.15f %22.15f] %25.15f" % (k+1,np.array(w_n),errval))
    print(k+1, np.array(w_n), errval)
    if errval <= eps:
        break
```

# 鸢尾花数据集的逻辑回归

* ## 用自编函数

```python
#加载相关包
from math import exp
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split

# 定义加载数据函数
def create_data():
    iris = load_iris()#加载鸢尾花数据集
    df = pd.DataFrame(iris.data, columns=iris.feature_names)#创建dataframe，列为鸢尾花数据集的特征
    df['label'] = iris.target#定义因变量y
    df.columns = ['sepal length', 'sepal width', 'petal length', 'petal width', 'label']#定义列名
    data = np.array(df.iloc[:100, [0,1,-1]])
    # print(data)
    return data[:,:2], data[:,-1]#返回前两列的x，以及y的列

X, y = create_data()#调用加载数据函数
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)#按照3:7将数据拆分成测试集和训练集

class LogisticReressionClassifier:#定义逻辑回归类
    def __init__(self, max_iter=200, learning_rate=0.01):#定义最大迭代次数为100，学习率为0.01
        self.max_iter = max_iter
        self.learning_rate = learning_rate

    def sigmoid(self, x):#定义sigmoid函数
        return 1 / (1 + exp(-x))

    def data_matrix(self, X):#定义数据矩阵
        data_mat = []
        for d in X:
            data_mat.append([1.0, *d])
        return data_mat

    def fit(self, X, y):
        # label = np.mat(y)
        data_mat = self.data_matrix(X)  # m*n
        self.weights = np.zeros((len(data_mat[0]), 1), dtype=np.float32)

        for iter_ in range(self.max_iter):
            for i in range(len(X)):
                result = self.sigmoid(np.dot(data_mat[i], self.weights))
                error = y[i] - result
                self.weights += self.learning_rate * error * np.transpose(
                    [data_mat[i]])
        print('LogisticRegression Model(learning_rate={},max_iter={})'.format(
            self.learning_rate, self.max_iter))

    # def f(self, x):
    #     return -(self.weights[0] + self.weights[1] * x) / self.weights[2]

    def score(self, X_test, y_test):#定义计算score的函数来评判模型好坏
        right = 0
        X_test = self.data_matrix(X_test)
        for x, y in zip(X_test, y_test):
            result = np.dot(x, self.weights)
            if (result > 0 and y == 1) or (result < 0 and y == 0):
                right += 1
        return right / len(X_test)#正确判断个数/测试集所有数据个数
```

```python
lr_clf = LogisticReressionClassifier()#加载逻辑回归模型
lr_clf.fit(X_train, y_train)#训练集数据带入模型
```

```python
lr_clf.score(X_test, y_test)#在测试集上计算模型的好坏
```

```python
x_ponits = np.arange(4, 8)#在4到8之间绘制散点
y_ = -(lr_clf.weights[1]*x_ponits + lr_clf.weights[0])/lr_clf.weights[2]#可视化分割直线即逻辑回归模型
plt.plot(x_ponits, y_)#画图

#lr_clf.show_graph()
plt.scatter(X[:50,0],X[:50,1], label='0')#绘制类别为0的点
plt.scatter(X[50:,0],X[50:,1], label='1')#绘制类别为1的点
plt.legend()#加上图例
```

# cohen kappa系数 & ROC曲线曲线
* ## cohen kappa系数：
$$
k=\frac{p_o-p_e}{1-p_e} .
$$
其中， $p_o$ 是每一类正确分类的样本数量之和除以总样本数，也就是总体分类精度。
假设每一类的真实样本个数分别为 $a 1, a 2, \ldots, \mathrm{aC}$ ，而预测出来的每一类的样本个数分别为 $\mathrm{b} 1, \mathrm{~b} 2, \ldots, \mathrm{bC}$
总样本个数为 $n$ ，则有: $\mathrm{pe}=\frac{\mathrm{a} 1 \times \mathrm{b} 1+\mathrm{a} 2 \times \mathrm{b} 2+\ldots+\mathrm{aC} \times \mathrm{bC}}{\mathrm{n} \times \mathrm{n}}$

kappa计算结果为-1到1，但通常kappa是落在 0 到1 间，可分为五组来表示不同级别的一致性：\
    0.0到0.20极低的一致性(slight)\
    0.21到0.40一般的一致性(fair)\
    0.41到0.60 中等的一致性(moderate)\
    0.61到0.80 高度的一致性(substantial)\
    0.81到1几乎完全一致(almost perfect)

分类问题中，最常见的评价指标是acc，它能够直接反映分正确的比例，同时计算非常简单。但是实际的分类问题种，各个类别的样本数量往往不太平衡。在这种不平衡数据集上如不加以调整，模型很容易偏向大类别而放弃小类别，此时整体acc挺高，但是部分类别完全不能被召回。这时需要一种能够惩罚模型的“偏向性”的指标来代替acc。而根据kappa的计算公式，越不平衡的混淆矩阵，acc越高，kappa值就越低，正好能够给“偏向性”强的模型打低分。


上图就是个混淆矩阵，
其中， $p_o=\frac{239+73+280}{664}=0.8916$.
$$
p_e=\frac{261 \times 276+103 \times 93+300 \times 295}{664 \times 664}=0.3883 .
$$
因此:
$$
k=\frac{0.8916-0.3883}{1-0.3883} .
$$
* ## ROC曲线
ROC的全称是Receiver Operating Characteristic Curve，中文名字叫“受试者工作特征曲线”。

该曲线的横坐标为假阳性率（False Positive Rate, FPR），N是真实负样本的个数，FP是N个负样本中被分类器预测为正样本的个数。
$$FPR=\frac{FP}{N} $$
纵坐标为真阳性率（True Positive Rate, TPR），P是真实正样本的个数，TP是P个正样本中被分类器预测为正样本的个数。
$$TPR=\frac{TP}{P} $$

绘制完成曲线后，就会对模型有一个定性的分析，如果要对模型进行量化的分析，此时需要引入一个新的概念，就是AUC（Area under roc Curve）面积，就是指ROC曲线下的面积大小，而计算AUC值只需要沿着ROC横轴做积分就可以了。真实场景中ROC曲线一般都会在这条直线的上方，所以AUC的取值一般在0.5到1之间。AUC的值越大，说明该模型的性能越好。

```python
from sklearn.model_selection import train_test_split
import statsmodels.api as sm
from patsy import dmatrices

train, test = train_test_split(#将数据按照3:7拆分成测试集和训练集，并且设置随机数种子
    titanic, test_size=0.3, stratify=titanic.Survived, random_state=0)#stratify是保持测试集与整个数据集里result的数据分类比例一致。

# 将分类变量转化为哑变量

y_train, X_train = dmatrices(#线性模型产生设计矩阵，模型为下面的线性模型，数据为train，返回形式是数据框
    'Survived ~ Class + Sex + Age', data=train, return_type='dataframe')

pd.options.display.max_columns = 10
X_train.head()#返回数据特征的前五行
```

```python
y_train.head()#返回数据label的前五行
```

```python
y_train = y_train.iloc[:,1]#y的训练集选择所有行
y_test, X_test = dmatrices('Survived ~ Class + Sex + Age', data=test, return_type='dataframe')#线性模型产生设计矩阵，模型为下面的线性模型，数据为train，返回形式是数据框
y_test = y_test.iloc[:,1]#y的测试集选择所有行
```

```python
model = sm.Logit(y_train, X_train)#带入statsmodels.api包的logit函数
results = model.fit()#拟合模型
```

```python
results.params#返回模型参数
```
